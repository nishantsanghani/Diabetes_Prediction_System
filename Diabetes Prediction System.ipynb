## Feature Engineering

```python
# Import zscore function from scipy.stats to detect outliers
from scipy.stats import zscore

# Calculate the z-score for each value in the DataFrame
# z-score measures how many standard deviations a value is from the mean
z = np.abs(zscore(df))

# Filter the DataFrame to keep only rows where all column values have z-score less than 3
# This effectively removes outliers (values more than 3 standard deviations from the mean)
dfi = df[(z < 3).all(axis=1)]

# Print the shape of the new DataFrame after removing outliers
print('dfi shape: ', dfi.shape)
```

```python
# Display the first 5 rows of the DataFrame
# Useful to quickly inspect the data after outlier removal or any preprocessing
df.head()
```

```python
# Create a new column 'BMI_Category' by categorizing BMI values into bins
# pd.cut() is used to segment continuous values into discrete intervals
# Bins are defined as:
#   0 - 18.5   : Underweight
#   18.5-24.9  : Normal weight
#   25-29.9    : Overweight
#   30-100     : Obese
# Labels [0,1,2,3] are assigned to these categories
# .astype(float) converts the category labels to float type
df['BMI_Category'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=[0,1,2,3]).astype(float)
```

```python
# Create a new column 'Glucose_Category' by categorizing Glucose values into bins
# pd.cut() is used to segment continuous glucose values into discrete intervals
# Bins are defined as:
#   0 - 70    : Low
#   71 - 99   : Normal
#   100 - 125 : Prediabetes
#   126 - 200 : High
# Labels [0,1,2,3] are assigned to these categories
# .astype(float) converts the category labels to float type
df['Glucose_Category'] = pd.cut(df['Glucose'], bins=[0, 70, 99, 125, 200], labels=[0,1,2,3]).astype(float)
```

```python
# Create a new column 'Age_Category' by categorizing Age values into bins
# pd.cut() is used to segment continuous age values into discrete intervals
# Bins are defined as:
#   20-30  : Young adult
#   31-40  : Adult
#   41-50  : Middle-aged
#   51-60  : Senior
#   61-100 : Elderly
# Labels [0,1,2,3,4] are assigned to these categories
# .astype(float) converts the category labels to float type
df['Age_Category'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 100], labels=[0,1,2,3,4]).astype(float)
```

```python
# Drop the 'BMI', 'Glucose', and 'Age' columns from the DataFrame
# This is typically done after converting these continuous variables into categorical versions 
# (e.g., 'BMI_Category', 'Glucose_Category', 'Age_Category'), making the original columns redundant
df = df.drop(['BMI','Glucose','Age'], axis=1, inplace=False)
```

## Train-Test Split

```python
# Select the feature columns for machine learning
# These include both original numerical columns and the newly created categorical columns
x = df[['Pregnancies', 'BloodPressure', 'SkinThickness', 'Insulin',
        'DiabetesPedigreeFunction',
        'Glucose_Category', 'Age_Category', 'BMI_Category']]

# Display the selected features
x
```

```python
# Select the target column for machine learning
# 'Outcome' indicates whether the individual has diabetes (1) or not (0)
y = df['Outcome']

# Display the target variable
y
```

```python
# Import train_test_split from scikit-learn to split the dataset
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# x_train, y_train will be used to train the model
# x_test, y_test will be used to evaluate the model
# test_size=0.20 means 20% of the data is reserved for testing
# random_state=42 ensures reproducibility of the split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

## Model Building

### Logistic Regression (with Scaling + CV)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Step 1: Standardize features
# Logistic Regression (and many ML algorithms) perform better when features are scaled
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Step 2: Initialize Logistic Regression model
# max_iter=1000  ensures convergence (default 100 may not be enough for complex data)
lr = LogisticRegression(max_iter=1000)

# Step 3: Perform 10-fold cross-validation
# cross_val_score automatically splits data into 10 folds and returns accuracy for each fold
scores = cross_val_score(lr, x_scaled, y, cv=10, scoring='accuracy')

# Step 4: Print mean accuracy across folds
print("Logistic Regression Accuracy:", scores.mean())
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Step 1: Initialize Decision Tree model
# random_state=42 ensures reproducibility (same splits each time)
dt = DecisionTreeClassifier(random_state=42)

# Step 2: Perform 10-fold cross-validation
# cross_val_score trains the model 10 times on different folds
scores = cross_val_score(dt, x, y, cv=10, scoring='accuracy')

# Step 3: Print the average accuracy
print("Decision Tree Accuracy:", scores.mean())
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Step 1: Initialize Random Forest
# Random Forest = an ensemble of multiple Decision Trees
# random_state=42  ensures reproducibility
rf = RandomForestClassifier(random_state=42)

# Step 2: Perform 10-fold cross-validation
# Random Forest will be trained on 10 different folds
scores = cross_val_score(rf, x, y, cv=10, scoring='accuracy')

# Step 3: Print average accuracy across folds
print("Random Forest Accuracy:", scores.mean())
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score

# Step 1: Initialize Gradient Boosting model
# Gradient Boosting = builds trees sequentially, each one fixing errors of the previous
gb = GradientBoostingClassifier(random_state=42)

# Step 2: Perform 10-fold cross-validation
# Evaluates the model on 10 different train-test splits
scores = cross_val_score(gb, x, y, cv=10, scoring='accuracy')

# Step 3: Print the average accuracy
print("Gradient Boosting Accuracy:", scores.mean())
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Step 1: Standardize features
# SVM is very sensitive to feature scale, so scaling is mandatory
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Step 2: Initialize SVM model
# By default, kernel = 'rbf' (non-linear)
svc = SVC()

# Step 3: Perform 10-fold cross-validation
# Cross-validation ensures performance is reliable across folds
scores = cross_val_score(svc, x_scaled, y, cv=10, scoring='accuracy')

# Step 4: Print average accuracy
print("SVM Accuracy:", scores.mean())
```

## Ensemble Models

### Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Step 1: Define base models
lr = LogisticRegression(max_iter=1000)               # Logistic Regression
rf = RandomForestClassifier(random_state=42)         # Random Forest
gb = GradientBoostingClassifier(random_state=42)     # Gradient Boosting
svc = SVC(probability=True)                          # Support Vector Machine
dt = DecisionTreeClassifier(random_state=42)         # Decision Tree

# Step 2: Build Voting Classifier
# voting='soft'  uses predicted probabilities (better than hard voting)
voting = VotingClassifier(
    estimators=[('lr', lr), ('rf', rf), ('gb', gb), ('svc', svc), ('dt', dt)],
    voting='soft'
)

# Step 3: Cross-validation for robust evaluation
print("Voting Classifier CV Accuracy:", cross_val_score(voting, x_scaled, y, cv=10).mean())
```

### Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Step 1: Define base learners (level-0 models)
estimators = [
    ("lr", LogisticRegression(max_iter=1000, random_state=42)),   # Logistic Regression
    ("rf", RandomForestClassifier(n_estimators=200, random_state=42)),  # Random Forest
    ("gb", GradientBoostingClassifier(n_estimators=200, random_state=42)),  # Gradient Boosting
    ("svc", SVC(probability=True, random_state=42))  # Support Vector Machine
]

# Step 2: Define stacking model
# final_estimator = RandomForest (meta-learner)
stack = StackingClassifier(
    estimators=estimators,
    final_estimator=RandomForestClassifier(n_estimators=200, random_state=42),
    cv=5
)

# Step 3: Train stacking classifier
stack.fit(x_train, y_train)

# Step 4: Predictions on test set
y_pred_stack = stack.predict(x_test)

# Step 5: Accuracy score
print("Stacking Accuracy:", accuracy_score(y_test, y_pred_stack))
```

## Advanced Models & Resampling

### XGBoost (with GridSearchCV)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier

# Step 1: Define the model
xgb = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)

# Step 2: Define hyperparameter grid
param_grid = {
    "n_estimators": [100, 200, 300],        # Number of trees
    "max_depth": [3, 5, 7],                # Depth of each tree
    "learning_rate": [0.01, 0.05, 0.1, 0.2],  # Step size shrinkage
    "subsample": [0.8, 1.0],               # % of samples used per tree
    "colsample_bytree": [0.8, 1.0]         # % of features used per tree
}

# Step 3: Grid Search with cross-validation
grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,              # 5-fold cross-validation
    scoring="accuracy",
    n_jobs=-1          # Use all CPU cores
)

# Step 4: Train with training data
grid.fit(x_train, y_train)

# Step 5: Get best model
best_xgb = grid.best_estimator_

# Step 6: Predict on test set
y_pred_xgb = best_xgb.predict(x_test)

# Step 7: Results
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Best Params:", grid.best_params_)
```

### SMOTE + SVM

### SMOTE + XGBoost with Hyperparameter Tuning (GridSearchCV & Stratified K-Fold CV)

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
import pandas as pd
import numpy as np

#  Train-test split with stratification
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

#  Apply SMOTE only on training set
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_train, y_train)

#  Base XGB model
xgb = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)

#  Hyperparameter grid
param_grid = {
    'n_estimators': [300, 500, 800],
    'max_depth': [4, 5, 6],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [1, 2, 5]
}

#  Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

#  GridSearchCV with AUC optimization
grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    verbose=1
)

#  Fit on resampled data
grid.fit(x_resampled, y_resampled)

#  Best model from grid search
best_xgb = grid.best_estimator_

#  Predictions on test data
y_pred = best_xgb.predict(x_test)
y_pred_proba = best_xgb.predict_proba(x_test)[:, 1]

#  Results
print("Best CV AUC:", grid.best_score_)
print("Best Params:", grid.best_params_)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Test ROC AUC:", roc_auc_score(y_test, y_pred_proba))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
